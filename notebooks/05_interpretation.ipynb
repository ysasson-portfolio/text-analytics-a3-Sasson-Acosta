### Why do differences exist?

Looking at the differences between the two models, there are overall differences between the groups based on the models because they vecotrize and tokenize the text in two completely different ways which will make the models produce different results. That is why Paris for the LDA model has 8 topics while the LMF model has 15 topics. 

If we are looking at the differences between the classes within the model, we can bring that down to a few diffferent reasons. The first reason is because the classes in this dataset are imbalanced. This means that there may not be as much information to utilize when we are building the topic models and can result in different matricies produced by CountVectorizer and TF-IDF. This will later bring different results when it comes to the optimal number of topics or the way that we fit the model to the dataset in the end.

Another reason that there could be a difference between the groups is because of differences in language from the region. Different areas of the world treat english differently. Examples of this could be slang, differentiated translation from other native languages, and other cultural differences depending on what part of the world you are from. This can change the way that the text is processed and add additional words to the dictionary rather than having a consistent list of vocabulary.
